## 1.3 度量选择

### 1.3.1 转移单位的选择 

该实验希望证明所增加的信息提示功能对之后整个免费试学转换为正式付费学习阶段是否有显著的正面作用，转移单位的选择有：Cookie，ID。  
因为ID只能参加一次免费试学，同时有些ID已经参加过免费试学甚至付费学习，则无法有效的测量该改进效果。  
所以这里转移单位使用Cookie更为合适。  
这里的Cookie数量的计算是按天计算的（在两个不同日期进行访问的同一个cookie 将计算两次）

### 1.3.2 不变指标的选择

相关的不变指标为如下：

| 不变指标      | 说明                    |
| :-------------- | :-------------------------------- |
| Cookie的数量   | 每天浏览该页的Cookie的数量      |
| Click的数量    | 每天送你点击免费试学按钮的Cookie的数量   |
| 点进概率（CTP） | Click的数量 / Cookie的数量      | 

每天浏览该页的Cookie数量，每天点击按钮的Cookie数量和其两者之比CTP，它们的行为都在看到改进的变化之前发生的（即免费数学筛选器是在点击按钮之后才跳出，只会对之后的行为产生影响），如果这三个指标在实验中出现了异常波动则说明可能是实验的数据收集出现问题，或者可能需要进步考虑对Cookie进行筛选以降低异常波动，其他造成该问题的可能原因都需要进步的研究

###  1.3.3 评估指标的选择

相关的评估指标为如下：  

| 评估指标      | 说明                         |
| :-------------- | :--------------------------------------- |
| 总转化率      | 进入免费学习的数量 / Cookie的数量      |
| 留存率       | 进入付费学习的数量 / 进入免费学习的数量  |
| 净转化率      | 进入付费学习的数量 / Cookie的数量      |

1. 总转化率：这指标反应的是：(点击按钮$\longrightarrow$免费试学)的比例，由于我们主要关心是否有更多人从免费试学环节顺利进入到付费学习环节，这两个环节的人数均以与总转化率有比例关系
2. 留存率： 这个指标反应的是：(免费试学$\longrightarrow$付费学习)的比例，是我们这次实验最直接相关的指标
3. 净转化率：这指标反应的是：(点击按钮$\longrightarrow$免费试学$\longrightarrow$付费学习)的比例，是上述两个指标的综合反应（$净转化率= 总转化率 \times 留存率$）。可以有效的增强实验的可信度并可以检验实验的正确性    

这三个指标的发生均在实验项目发生之后，并且其改进对这些指标的影响层次各有不同，不可作为不变指标   
<br/>    $\star \;$用户ID数量不作为评估度量的原因：因为用户ID数量没有归一化，由于实验组和对照组的cookie数量不一定相同，也就是说两组中用户ID数量不同可能是由于实验的影响，也可能是由于两组cookie的不同。若两户ID数量存在差异，我们无法区分是由哪个原因造成的。



## 1.4 试验希望取得的结果

|          | 总转化率   | 留存率     | 净转化率	   | 
| -------------------: | :--------: | :-------: | :----------: |
| 实际显著($D_{min}$)   | 0.01       | 0.01      | 0.0075       |

若能达到我们在$1.2$中所期望的结果，则以上指标应为一下变化：  
1. 不变变量不应出现任何统计显著。
2. 总转化率应会产生反向变化（因为当学员填入小于5小时，学员会被导入学习免费课程页面，若要继续参加则需再自行返回该页面参与免费试学）  
同时我们希望总转化率的下降不要超过这里设定的实际显著要求：$1\%$。
3. 留存率应发生正向变化（因为该实验所希望达到的目的就是减少在免费试学环节因时间分配不足而退出的学员）。
4. 对于净转化率，我们期望该指标至少不降低。  

## 1.5 计算标准偏差（变异性）

对于该实验的评估指标计算其标准偏差，所有信息均通过在实验测试前这些指标已有的标准值进行计算，首先先导入标准值：

```python
baseline_df = pd.read_excel("Final+Project+Baseline+Values.xlsx",0,header=None,names=["Metric","Value"]);baseline_df #导入标准值表
```

在这里首先假设每天浏览网页的Cookie数量为5000，再次基础上计算各评估指标的标准偏差：

```python
def std_analytic(baseline_df):# 计算每个暂定评估指标的标准偏差（基于每天唯一Cookie为5000）
    N_click=baseline_df.iloc[1]["Value"];N_enroll=baseline_df.iloc[2]["Value"]
    P_enroll = baseline_df.iloc[4]["Value"];P_Pay_enroll= baseline_df.iloc[5]["Value"]
    P_Pay_click=baseline_df.iloc[6]["Value"]
    Index=[];M=[]
    Gross_conversion_se =  np.sqrt(P_enroll*(1-P_enroll)/N_click)*(np.sqrt(1/float(5000))/np.sqrt(1/float(40000)))
    Retention_se = np.sqrt(P_Pay_enroll*(1-P_Pay_enroll)/N_enroll)*(np.sqrt(1/float(5000))/np.sqrt(1/float(40000)))
    Net_conversion_se = np.sqrt(P_Pay_click*(1-P_Pay_click)/N_click)*(np.sqrt(1/float(5000))/np.sqrt(1/float(40000)))
    M.extend([Gross_conversion_se,Retention_se,Net_conversion_se]);Index.extend(["总转化率","留存率","净转化率"])
    data=list(zip(Index,M))
    df=pd.DataFrame(data,columns=["Metric","SE"])
    return  df
std_analytic(baseline_df)
```

在本次实验设计的分组单位：Cookie ，其中：
1. 总转化率 $ \& $ 净转化率的分析单位均为：Cookie，所以其分析变异性与经验变异性相匹配
2. 留存率的分析单位：进入免费试学的数量(即ID数量)，所以其分析变异性与经验变异性相不匹配，实际的偏差采用经验所得的变异性为宜。

## 1.6 计算样本容量

首先保留上述三个评估指标，其中我们看到三个指标的两两之间存在很大的关联，在这里如果使用Bonferroni 校正会导致实验过分保守，不利于观察到我们所感兴趣的变化。以下在给定的要求下并通过[在线A/B Test样本容量计算器](http://www.evanmiller.org/ab-testing/sample-size.html)，相关数据仍来自该标准值表 ,然后通过 $ \{ {Value_{Pageview} \div Value_{评估指标样本单位}}  \times 2 \} $各评估指标的样本容量要求都调整为每天的总Cookie数量（因为数据两要分给实验组和观察组所以需要翻倍）

| 评估指标   | $\alpha$ | 敏感度（1-$\beta $） | 实际显著($D_{min}$) | 所需样本数量 | 单位 | 转换后Pageviews的数量 |
| ----------: | :----: | :--------------: | :-------------: | :----------: | -------: | :----------: |
| 总转化率   | 0.05  | 0.8          | 0.01         | 25,835      | Clicks   | 645,875    |
| 留存率    | 0.05   | 0.8          | 0.01        | 39,115      | Enrollments | 4,741,212  |
| 净转化率   | 0.05  | 0.8          | 0.0075        | 27,413     | Clicks    | 685,325    |


其中我们看到总转化率和净转化率对浏览网页的总Cookie数量都比较接近，而留存率对样本容量的要求比其他两个指标的要求高了7倍。  
这对数量和实验所需时间的要求就大大提高了，同时留存率的指标与其他两者的指标关联度也较大。在这里选择将留存率从评估指标中剔除  
因此在剩下的两个指标的要求中选择较大的一个：即样本所需的Pageviews数量：$685,325 $ Cookies

## 2.2 完整性检查

首先在进行实验数据分析之前先进行完整性检查，在设计实验时选择的不变指标应在实验测试中不受影响，即指标在实验组和对照组的大小应该没有显著差别

### 2.2.1 Cookie 数量检查

在这里选用$P= PAGEVIEW\_cont \div PAGEVIEW\_exp$ 作用检查对象，$P$在两个组之间的分布服从二项分布，因这里$N$很大，可以用正态分布近似， $P\thicksim N(0.5,SE^2 )$，其中 $SE = \sqrt{\frac{P \cdot (1-P)}{N}}$,从而得到该期望值的置信区间：

```python
cookie_sanity = prop.binom_test_reject_interval(0.5,690203) # cookie 的完整性检验
P_cookie= float(cont_df["Pageviews"].sum()) / float(exp_df["Pageviews"].sum()+cont_df["Pageviews"].sum())
print("P_Cookie置信区间:", [int(i)/float(690203) for i in cookie_sanity] )
print("Cookie比值:","     ",P_cookie)
```

实验中检查对象为 $0.5006$，落在置信区间$[\; 0.4988 \; , \;  0.5012 \; ]$内，该不变指标满足完整性检查

### 2.2.2 Click 数量检查

Click数量的方法与Cookie数量的方法一样，计算过程如下：

```python
click_sanity=prop.binom_test_reject_interval(0.5,56703) # click 的完整性检验
P_click= float(cont_df["Clicks"].sum()) / float(exp_df["Clicks"].sum()+cont_df["Clicks"].sum())
print("P_Click置信区间:",[int(i)/float(56703) for i in click_sanity])
print("Click比值:","      ",P_click)
```

实验中检查对象为 $0.5005$，落在置信区间$[\; 0.4959 \; , \;  0.5041 \; ]$内，该不变指标满足完整性检查

### 2.2.3 点进概率（CTP）检查

这里点进概率$CTP = {SUM_{Click} \div SUM_{Pageviews}} $，通过假设检验的方法验证两个组中的CTP是否有显著差异：
$$d= CTP_{exp} - CTP_{cont}\; \,\; \,\; \, H_0\colon \; \,  d= 0 \; \,  H_1\colon \; \,  d \neq 0 $$  

在$H_0\colon \; \,  d= 0$成立的假设下，有$CTP_{exp} = CTP_{cont}=CTP_{pool}$, 其中:  

$$CTP_{pool} = {{CLICK_{cont}+CLICK_{exp} \over {PAGEVIEW_{cont}+PAGEVIEW_{exp}}}}$$  
$$SE_{pool}=\sqrt{CTP_{pool} \cdot (1-CTP_{pool}) \cdot (\frac{1}{PAGEVIEW_{cont}}+\frac{1}{PAGEVIEW_{exp}})}$$  
所以有：$ d\thicksim N(\,0\,,\,SE^2 \,)$，计算$\hat{d}$是否在$d$的置信区间内判断是否显著有差异

```python
p_ctp = summary.iloc[1] / summary.iloc[0] 
d_hat = p_ctp.iloc[1]-p_ctp.iloc[0]
print("CTP_exp:    %.5f" % (p_ctp.iloc[1]))
print("CTP_cont:   %.5f" % (p_ctp.iloc[0]))
print("d_hat:    %.4f" % (d_hat))
p_pool=(summary.iloc[1].sum() / summary.iloc[0].sum())
se_pool =np.sqrt(p_pool*(1-p_pool)*(1/summary.iloc[0][0]+1/summary.iloc[0][1]))
margin=sp.stats.norm.ppf(0.975)*se_pool
ci=(-margin,margin)
print("CTP_pool:   %.4f" % (p_pool))
print("SE_pool:  %.4f" % (se_pool))
print("d的置信区间: [ %.4f , %.4f ] " % (-margin,margin))
```

$d$的置信区间为$[\,-0.013\,,\,0.013\,]$，$\hat{d}=0.0001$落在该置信区间内，则CTP满足完整性检查

## 2.3 评估指标分析

### 2.3.1 实验数据进步整理

对评估指标的分析是要是判断该指标是否统计显著和实际显著，通过实验所得数据构造该指标估计值的置信区间去判断具有这两个显著特性。  
因为实验设计中评估指标主要观察未参加过课程的人向免费试学和付费学习的转化比率，其中付费学习必须在14天后才能观察到，  
因此实验数据中Enrollments和payments中的数据与Click和Pageviews的数据差14天，这里使用队列的方式只考虑与Enrollments和payments对应的实验数据，因此对实验数据进行进步整理

```python
cont_df1=cont_df.dropna() #在分析总/净转化率时以enrollments和payments为以准保留数据
exp_df1=exp_df.dropna()
print("control group info:")
cont_df1.info()
```

```python
print("expriment group info:")
exp_df1.info()
```

对以Enrollments和payments调整的实验数据进行汇总：

```python
ab1 = cont_df1[["Pageviews","Clicks","Enrollments","Payments"]].sum()
cd1 =exp_df1[["Pageviews","Clicks","Enrollments","Payments"]].sum()
summary1=pd.merge(ab1.reset_index(),cd1.reset_index(),on="index")
summary1.rename(columns={'index':'SUM', '0_x':'Control Group', '0_y':'Experiment Group'}, inplace = True)
summary1.index=summary1['SUM'].values;del summary1["SUM"]
summary1.columns=[["SUM_trimed","SUM_trimed"],["Control Group",'Experiment Group']];
summary1
```

```python
def conversion_analysis(string): # 自定义一个函数来计算总/净转化率，返回转化率估计值和其置信区间
    if string=="gross":
        numerator=2;name="GrossConversion"
    elif string=="net":
        numerator=3;name="NetConversion"
    p_ctp1 = summary1.iloc[numerator] / summary1.iloc[1]
    Gconv_hat = p_ctp1.iloc[1]-p_ctp1.iloc[0]
    print("%s_exp:    %.5f" % (name,p_ctp1.iloc[1]))
    print("%s_cont:   %.5f" % (name,p_ctp1.iloc[0]))
    print("%s_hat:    %.4f" % (name,Gconv_hat))
    p_pool1=(summary1.iloc[numerator].sum() / summary1.iloc[1].sum())
    se_pool1 =np.sqrt(p_pool1*(1-p_pool1)*(1/summary1.iloc[1][0]+1/summary1.iloc[1][1]))
    margin1=sp.stats.norm.ppf(0.975)*se_pool1
    print("%s_hat的置信区间： （%.4f，%.4f）" % (name,Gconv_hat-margin1,Gconv_hat+margin1))
```

### 2.3.2 总转化率分析

总转化率的验证方法与CRT的方法相同，只是在计算CRT的时候因为我们已经假设$d=0$，所以用数据去验证这个假设。  
而在这里总转化率的分析中我们并不知道$d$的期望值，因此我们需要用实验数据 $\hat{d}$ 的置信区间去估计$d$的真值范围，从而去判断是否满足统计显著和实际显著的特征。

```python
conversion_analysis("gross")
```

总转化率的实际显著要求为$[\,-0.01\,,\,0.01\,]$。可以看到$\hat{GrossConversion}$的置信区间：$[\,-0.291\,,\,-0.0120\,]$,很显然，总转化率具有统计显著 & 实际显著

```python
x=[-0.0291,-0.0120];y=[2,2] #画出总转化率分析的图形示意
y_err=[0.2,0.2]
plt.figure(figsize=(6,3))
plt.errorbar(x,y, yerr=y_err)
plt.scatter(-0.0206,2,linewidth=2)
plt.xlim(-0.03,0.03);plt.xlabel("Difference");plt.xticks([-0.02,-0.01,0,0.01,0.02])
plt.vlines(-0.01, 0, 4,linestyles="--",colors="gray");plt.vlines(0.01, 0, 4,linestyles="--",colors="gray")
plt.vlines(0, 0, 4,colors="k")
plt.annotate("GrossConversion CI",xy=[-0.03,1.3])
plt.yticks([]);ax=plt.gca()
ax.spines["right"].set_color("none");ax.spines["top"].set_color("none")
ax.spines["left"].set_color("none");ax.spines["bottom"].set_color("none")
```

### 2.3.3 净转化率分析

净转化率的分析方法与总转化率的分析方法相同。

```python
conversion_analysis("net")
```

## 2.4 符号检验

最后用符号检验验证下评估指标。  
通过观察每组数据中实验组的评估指标大于观察组的评估指标的次数，去检验该实验所产生的变化是否显著。

### 2.4.1 总转化率符号检验

```python
G_sign=exp_df1["Enrollments"]/exp_df1["Clicks"]>cont_df1["Enrollments"]/cont_df1["Clicks"] # 算出实验组指标>观察组指标的数据 ---总转化率
print("实验中总共有%d组数据，其中有%d次实验组的指标>观察组的指标" % (G_sign.count(),G_sign.sum()))
p1 =prop.binom_test(G_sign.sum(),G_sign.count()) #总转化率的符号检验
print("总转化率的符号检验P值：%.4f" % (p1))
```

在实验中有23组数据，其中有4组实验组的总转化率>观察组的总转化率。P值：$0.0026 << \alpha$，与之前分析相符

### 2.4.2 净转化率符号检验

```python
N_sign=exp_df1["Payments"]/exp_df1["Clicks"]>cont_df1["Payments"]/cont_df1["Clicks"] # 算出实验组指标>观察组指标的数据 ---净转化率
print("实验中总共有%d组数据，其中有%d次实验组的指标>观察组的指标" % (N_sign.count(),N_sign.sum()))
p2=prop.binom_test(N_sign.sum(),N_sign.count()) #净转化率的符号检验
print("净转化率的符号检验P值：%.4f" % (p2))
```

在实验中有23组数据，其中有10组实验组的总转化率>观察组的总转化率。P值：$0.6776 >> \alpha$，与之前分析相符

## 3.1 决策建议

首先将实验所得分析集中罗列：  

| 评估指标  |  统计显著    | 实际显著      |  实际显著变化方向  |
| -------: | :-----------: | :---------: | :-------: |
| 总转化率  | $\checkmark$ | $\checkmark$ | Negitive  |
| 净转化率  | $\times$     | $\times$     | Negative     |  
1. 对于总转化率，虽然它同时具有统计显著和实际显著，然而其变化的方向反向的，即若采用弹出信息功能不仅不会使得更多的人参与到免费试学，反而会使得参加免费试学的人数显著减少。
2. 对于净转化率，由于其不具有统计显著，同时其估计值的置信区间存在部分区域会超出实际显著的要求，然后其方向也为反向(即在一定概率下会减少付费学习的比例)。   
  
综上，根据实验的信息可以给出的决策建议：
